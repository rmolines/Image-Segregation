Neste projeto tivemos de paralelizar um código pronto de segmentação de imagens através do algoritimo MSSP - Multiple Source Shortest Path. ### Segmentação de Imagens - MSSP

A ideia de segmentar imagens é separar algum objeto do resto da imagem. No nosso projeto, fizemos isso com o algoritmo MSSP - Multiple Source Shortest Path, ou caminho mais curto de múltiplas fontes. A ideia desse algoritimo é tratar a imagem como um grafo, e assim, atribuindo pesos às arestas que ligam os vértices desse grafo, poder traçar a menor rota entre todos os demais vértices e os nós fonte. Faz-se então isso para ambos os nós da parte da imagem que se deseja permanecer, e com os nós da parte que se deseja retirar. Esses nós representam pontos na imagem que são selecionados pelo usuário. 


### Especificações

Para este projeto, tínhamos à disposição um cluster de máquivas virtuais rodando na AWS. Com a criação de um Network File System, pudemos compartilhar uma pasta que continha o nosso crawler, e assim usufruir de todas as máquinas no cluster.

### Distribuição

O programa usa Cmake para criar a build. Basta ter as bibliotecas instaladas e rodar 'cmake . && make' no diretório raíz para criar o executáveis. Na máquina virtual que está na AWS, isso não foi possível, por isso foi dado junto ao programa também um Makefile bem simples que compila as versões sequencial e paralela.
Para rodar o programa, basta rodar './gpu [arquivo de entrada] [arquivo de saída]'. Para a versão sequencial, basta trocar 'gpu' por 'seq'. Com isso, o usuário deve então digitar o número de nós para a imagem da frente, seguido pelo número de nós do plano de fundo. Então digita-se os pares de coordenadas para os os nós da frente, em seguida para os nós de trás. Ao final do programa, será escrito na saída padrão os tempos do programa, além da imagem resultante no nome do arquivo de saída.

### Testes

Na teoria os tempos utilizando o poder de processamento da GPU deverem ser muito mais rápidos que usando apenas a CPU, isso porque as GPUs tem a característica de terem várias threads mais simples e mais lentas, em contraste a CPUs que têm poucas threads muito poderosas e muito rápidas. Apesar disso, o que observamos nos resultados dos testes foi que o programa rodando em CUDA foi extremamente mais lento que o rodando em CPU.
